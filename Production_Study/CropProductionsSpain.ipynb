## Strawberry country data

```python
import pandas as pd
pd.set_option('display.max_columns',None) ## Display all columns in pandas dataframe  
pd.set_option('display.max_rows',None) ## Display all rows in pandas dataframe 

sheet_name_1117='13.6.30.1'
sheet_name_18='7.6.30.1'
columns_skip=[1,2,3,4,5,6,7,8]
columns_ok=None
file_names_year = [  ['AE_2011_13.xls',2011,sheet_name_1117,columns_ok]\
                    ,['AE_2012_13.xls',2012,sheet_name_1117,columns_ok]\
                    ,['AE_2013_13.xls',2013,sheet_name_1117,columns_ok]\
                    ,['AE_2014_13.xlsx',2014,sheet_name_1117,columns_ok]\
                    ,['AE15-C13.xlsx',2015,sheet_name_1117,columns_ok]\
                    ,['AE16-C13.xlsx',2016,sheet_name_1117,columns_ok]\
                    ,['AE17-C13.xlsx',2017,sheet_name_1117,columns_ok]\
                    ,['AE18-C07.xlsx',2018,sheet_name_18,columns_ok]\
                  ]
crop_productions_es_strw = pd.DataFrame()
for i in range(0,len(file_names_year)):
    crop_prod_yr=pd.read_excel(file_names_year[i][0]\
                               ,sheet_name=file_names_year[i][2]\
                               ,header=6\
                               ,skiprows=[7,8]\
                               ,usecols=file_names_year[i][3])
    crop_prod_yr['YearAnalysis']=file_names_year[i][1]
    crop_prod_yr.columns = ['Year','Surface','Yield','Production','AvgPrice','Amount','YearAnalysis']
    crop_productions_es_strw=crop_productions_es_strw.append(crop_prod_yr)
    

crop_productions_es_strw.dropna(axis=1,inplace=True)
crop_productions_es_strw.drop_duplicates(subset=['Year','Surface','Yield','Production','AvgPrice','Amount'],inplace=True)
crop_productions_es_strw
```

```python
import seaborn as sns
print(sns.heatmap(crop_productions_es_strw.corr()))
```

```python
crop_productions_es_strw.groupby('Year')['Production'].agg('sum')
```

## Avocado country data

```python
import pandas as pd
pd.set_option('display.max_columns',None) ## Display all columns in pandas dataframe  
pd.set_option('display.max_rows',None) ## Display all rows in pandas dataframe 

sheet_name_1117='13.9.15.1.'
sheet_name_18='7.9.15.1.'
columns_skip=[1,2,3,4,5,6,7,8]
columns_ok=None
file_names_year = [ ['AE_2011_13.xls',2011,sheet_name_1117,columns_ok]\
                    ,['AE_2012_13.xls',2012,sheet_name_1117,columns_ok]\
                    ,['AE_2013_13.xls',2013,sheet_name_1117,columns_ok]\
                    ,['AE_2014_13.xlsx',2014,sheet_name_1117,columns_ok]\
                    ,['AE15-C13.xlsx',2015,sheet_name_1117,columns_skip]\
                    ,['AE16-C13.xlsx',2016,sheet_name_1117,columns_skip]\
                    ,['AE17-C13.xlsx',2017,sheet_name_1117,columns_skip]\
                    ,['AE18-C07.xlsx',2018,sheet_name_18,columns_ok]\
                  ]
crop_productions_es_strw = pd.DataFrame()
for i in range(0,len(file_names_year)):
    crop_prod_yr=pd.read_excel(file_names_year[i][0],\
                                sheet_name=file_names_year[i][2],\
                                header=6,\
                                skiprows=[7,8],\
                                skipcolumns=1,\
                                usecols=file_names_year[i][3])
    crop_prod_yr['YearAnalysis']=file_names_year[i][1]
    crop_prod_yr.columns = ['Year','SurfaceTotal','SurfaceProduction','Trees','Yield','Production','AvgPrice','Amount','YearAnalysis']
    crop_productions_es_avo=crop_productions_es_avo.append(crop_prod_yr)
    

crop_productions_es_avo.dropna(axis=1,inplace=True)
crop_productions_es_avo.drop_duplicates(subset=['Year','SurfaceTotal','SurfaceProduction','Trees','Yield','Production','AvgPrice','Amount']\
                                        ,inplace=True)
crop_productions_es_avo.sort_values(by='Year',ascending=True)
```

```python
import seaborn as sns
print(sns.heatmap(crop_productions_es_avo.corr()))
```

## Strawberry region data

```python
import pandas as pd
pd.set_option('display.max_columns',None) ## Display all columns in pandas dataframe  
pd.set_option('display.max_rows',None) ## Display all rows in pandas dataframe 

sheet_name_1117='13.6.30.2'
sheet_name_18='7.6.30.2'
columns_skip=[1,2,3,4,5,6,7,8]
columns_ok=None
file_names_year = [ ['AE_2011_13.xls',2011,sheet_name_1117,columns_ok]\
                    ,['AE_2012_13.xls',2012,sheet_name_1117,columns_ok]\
                    ,['AE_2013_13.xls',2013,sheet_name_1117,columns_ok]\
                    ,['AE_2014_13.xlsx',2014,sheet_name_1117,columns_ok]\
                    ,['AE15-C13.xlsx',2015,sheet_name_1117,columns_ok]\
                    ,['AE16-C13.xlsx',2016,sheet_name_1117,columns_ok]\
                    ,['AE17-C13.xlsx',2017,sheet_name_1117,columns_ok]\
                    ,['AE18-C07.xlsx',2018,sheet_name_18,columns_ok]\
                  ]
crop_productions_es_strw_reg = pd.DataFrame()
for i in range(0,len(file_names_year)):
    crop_prod_yr=pd.read_excel(file_names_year[i][0]\
                               ,sheet_name=file_names_year[i][2]\
                               ,header=5\
                               ,na_values=['','-','']
                               ,usecols=file_names_year[i][3])
    crop_prod_yr['Year']=file_names_year[i][1]-1
    crop_prod_yr['YearAnalysis']=file_names_year[i][1]
    crop_prod_yr.columns = ['Region','SurfaceDry','SurfaceIrOpen','SurfaceIrProtect','SurfaceTotal','YieldDry','YieldIrOpen','YeldIrProtect','Production','Year','YearAnalysis']
    crop_productions_es_strw_reg=crop_productions_es_strw_reg.append(crop_prod_yr)

crop_productions_es_strw_reg=crop_productions_es_strw_reg.reset_index()

# Fill NaNs
crop_productions_es_strw_reg['Production'].fillna(value=0,inplace=True)
crop_productions_es_strw_reg['SurfaceIrOpen'].fillna(value=0,inplace=True)
crop_productions_es_strw_reg['SurfaceIrProtect'].fillna(value=0,inplace=True)
crop_productions_es_strw_reg['YieldIrOpen'].fillna(value=0,inplace=True)
crop_productions_es_strw_reg['YeldIrProtect'].fillna(value=0,inplace=True)

# Change column types
crop_productions_es_strw_reg.SurfaceIrOpen = crop_productions_es_strw_reg.SurfaceIrOpen.astype('float64')
crop_productions_es_strw_reg.SurfaceIrProtect = crop_productions_es_strw_reg.SurfaceIrProtect.astype('float64')
crop_productions_es_strw_reg.YieldIrOpen = crop_productions_es_strw_reg.YieldIrOpen.astype('float64')
crop_productions_es_strw_reg.YeldIrProtect = crop_productions_es_strw_reg.YeldIrProtect.astype('float64')

# Delete blank rows
crop_productions_es_strw_reg.dropna(axis=0,subset=['Region'],inplace=True)

# https://stackoverflow.com/questions/13851535/delete-rows-from-a-pandas-dataframe-based-on-a-conditional-expression-involving
# crop_productions_es_strw_reg.drop(crop_productions_es_strw_reg[crop_productions_es_strw_reg['Region'].str.isupper()].index,inplace=True)

# Delete regions grouping to avoid duplications
regions_to_delete=[ 'GALICIA',\
                    'PAS VASCO',\
                    'ARAGN',\
                    'CATALUA',\
                    'CASTILLA Y LEN',\
                    'CASTILLA-LA MANCHA',\
                    'C. VALENCIANA',\
                    'EXTREMADURA',\
                    'ANDALUCA',\
                    'CANARIAS',\
                    'ESPAA']
crop_productions_es_strw_reg['Region']=crop_productions_es_strw_reg['Region'].str.strip()
crop_productions_es_strw_reg.drop(crop_productions_es_strw_reg[crop_productions_es_strw_reg.Region.isin(regions_to_delete)].index,inplace=True)
crop_productions_es_strw_reg.drop('index',axis=1,inplace=True)

#crop_productions_es_strw_reg.to_excel('StrawberryRegions.xlsx')
```

```python
import seaborn as sns
print(sns.heatmap(crop_productions_es_strw_reg.corr()))
```
